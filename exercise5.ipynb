```python
import numpy as np
import matplotlib.pyplot as plt
from least_squares import *
```

```python
#Ex1 c)
xs = np.array([
    [0, 1],
    [0, 1],
    [1, 0],
    [1, 0],
    [1, 2]])

ys = np.array([3, 2, 3, 4, 6])
fs = lambda x: np.insert(x, 0, 1)

# print(np.linalg.inv(xs.T @ xs) @ xs.T @ ys)

A = make_fun(fs, xs, (xs.shape[0], xs.shape[1] + 1))
t = least_squares(A, ys)

print(t)
```

```python
#Ex 1 d)
n = 1000
m = 10

phi = lambda x: np.array([np.sin(x * (k+1)) for k in range(m)])

xs = gendata(n)
ys = xs / 2
A = make_fun(phi, xs, (n, m))
t = least_squares(A, ys)
print(t)

xvals = np.linspace(-np.pi * 6, np.pi * 6, 1000)
plt.plot(xvals, A @ t)
plt.show()
```
